```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xarray as xr
import numpy as np
import pandas as pd
import xarray as xr
import scipy.stats
import tensorflow as tf
from glob import glob
from tensorflow import keras
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import *
from tensorflow.keras import Sequential
from utils import * 

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
%load_ext autoreload
%autoreload 2


# train
sithick_train = xr.open_dataset('./Potential Dataset/ice thickness-1/sithick.nc')
siconc_train = xr.open_dataset('./Potential Dataset/ice thickness-1/siconc.nc')
so_train = xr.open_dataset('./Potential Dataset/ice thickness-1/so.nc')
thetao_train = xr.open_dataset('./Potential Dataset/ice thickness-1/thetao.nc')
usi_train = xr.open_dataset('./Potential Dataset/ice thickness-1/usi.nc')
vsi_train = xr.open_dataset('./Potential Dataset/ice thickness-1/vsi.nc')
bottomT_train = xr.open_dataset('./Potential Dataset/ice thickness-1/bottomT.nc')


# test
sithick_test = xr.open_dataset('./Potential Dataset/ice thickness-1/sithick_test.nc')
siconc_test = xr.open_dataset('./Potential Dataset/ice thickness-1/siconc_test.nc')
so_test = xr.open_dataset('./Potential Dataset/ice thickness-1/so_test.nc')
thetao_test = xr.open_dataset('./Potential Dataset/ice thickness-1/thetao_test.nc')
usi_test = xr.open_dataset('./Potential Dataset/ice thickness-1/usi_test.nc')
vsi_test = xr.open_dataset('./Potential Dataset/ice thickness-1/vsi_test.nc')
bottomT_test = xr.open_dataset('./Potential Dataset/ice thickness-1/bottomT_test.nc')
```

```python
sithick_train["sithick"].mean(dim=["latitude", "longitude"])
```

```python
def preprocess_dataset(dataset, dimensions_to_average=None):
    """
    Averages the dataset over the specified dimensions, if provided.
    """
    if dimensions_to_average:
        return dataset.mean(dim=dimensions_to_average)
    else:
        return dataset

# Preprocessing training datasets
siconc_train_avg = preprocess_dataset(siconc_train["siconc"], ["latitude", "longitude"])
so_train_avg = preprocess_dataset(so_train["so"], ["latitude", "longitude", "depth"])
thetao_train_avg = preprocess_dataset(thetao_train["thetao"], ["latitude", "longitude", "depth"])
usi_train_avg = preprocess_dataset(usi_train["usi"], ["latitude", "longitude"])
vsi_train_avg = preprocess_dataset(vsi_train["vsi"], ["latitude", "longitude"])
bottomT_train_avg = preprocess_dataset(bottomT_train["bottomT"], ["latitude", "longitude"])
sithick_train_avg = preprocess_dataset(sithick_train["sithick"], ["latitude", "longitude"])

# Preprocessing testing datasets
siconc_test_avg = preprocess_dataset(siconc_test["siconc"], ["latitude", "longitude"])
so_test_avg = preprocess_dataset(so_test["so"], ["latitude", "longitude", "depth"])
thetao_test_avg = preprocess_dataset(thetao_test["thetao"], ["latitude", "longitude", "depth"])
usi_test_avg = preprocess_dataset(usi_test["usi"], ["latitude", "longitude"])
vsi_test_avg = preprocess_dataset(vsi_test["vsi"], ["latitude", "longitude"])
bottomT_test_avg = preprocess_dataset(bottomT_test["bottomT"], ["latitude", "longitude"])
sithick_test_avg = preprocess_dataset(sithick_test["sithick"], ["latitude", "longitude"])

```

```python
slider = 5  # Slider value for the input data window

# Function to create datasets
def create_dataset(*datasets, slider):
    combined_data = []
    for i in range(len(datasets[0]) - slider):
        data_slice = [dataset[i:(i + slider)].values for dataset in datasets]
        combined_slice = np.concatenate(data_slice, axis=-1)
        combined_data.append(combined_slice)
    return np.array(combined_data)

# Creating input datasets
X_train = create_dataset(siconc_train_avg, so_train_avg, thetao_train_avg, 
                         usi_train_avg, vsi_train_avg, bottomT_train_avg, 
                         slider=slider)

X_test = create_dataset(siconc_test_avg, so_test_avg, thetao_test_avg, 
                        usi_test_avg, vsi_test_avg, bottomT_test_avg, 
                        slider=slider)

# Assuming one-step-ahead prediction for targets
Y_train = sithick_train_avg.values[slider:]
Y_test = sithick_test_avg.values[slider:]

# Print shapes of datasets

X_train_reshaped = X_train.reshape(X_train.shape[0], -1) 
X_train_df = pd.DataFrame(X_train_reshaped)

Y_train_reshaped = Y_train.reshape(Y_train.shape[0], -1) 
Y_train_df = pd.DataFrame(Y_train_reshaped)

X_test_reshaped = X_test.reshape(X_test.shape[0], -1) 
X_test_df = pd.DataFrame(X_test_reshaped)

Y_test_reshaped = Y_test.reshape(Y_test.shape[0], -1) 
Y_test_df = pd.DataFrame(Y_test_reshaped)

X_train = X_train_df.interpolate(method='linear', axis=0)
Y_train = Y_train_df.interpolate(method='linear', axis=0)
X_test = X_test_df.interpolate(method='linear', axis=0)
Y_test = Y_test_df.interpolate(method='linear', axis=0)

X_train_missing = X_train_df.isna().any().any()
Y_train_missing = Y_train_df.isna().any().any()
X_test_missing = X_test_df.isna().any().any()

# Convert DataFrames to NumPy arrays
X_train = X_train.to_numpy()
Y_train = Y_train.to_numpy()
X_test = X_test.to_numpy()
Y_test = Y_test.to_numpy()

# Number of time steps (the length of your sliding window)
num_timesteps = slider

# Number of features (assuming all datasets have the same number of features)
num_features = X_train.shape[1] // num_timesteps

# Reshaping X_train and X_test
X_train = X_train.reshape(-1, num_timesteps, num_features)
X_test = X_test.reshape(-1, num_timesteps, num_features)

# For Y_train and Y_test, assuming you are predicting a single feature at a single time step
Y_train = Y_train.reshape(-1, 1, 1)  # This might change based on your model's requirements
#Y_test = Y_test.reshape(-1, 1, 1)   # This might change based on your model's requirements


print(X_train.shape, Y_train.shape, X_test.shape, Y_test.shape)
```

```python
# set hyperparameters
n_neuron       = 64
activation     = 'relu'
num_epochs     = 60
learning_rate  = 0.05
minibatch_size = 64
model_num      = 1
```

```python
lstm_model = Sequential()
lstm_model.add(LSTM(n_neuron,input_shape=(X_train.shape[1],X_train.shape[2]),
               return_sequences=True,activation=activation))
lstm_model.add(LSTM(n_neuron,return_sequences=False,
               activation=activation))
lstm_model.add(Dense(n_neuron,activation=activation))
lstm_model.add(Dense(Y_train.shape[-1],activation='linear')) 
print(Y_train.shape[-1])
lstm_model.compile(loss='mse',optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate))

lstm_model.summary()

print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=20)

history = lstm_model.fit(X_train, Y_train, 
                        batch_size = minibatch_size,
                        epochs = num_epochs,
                        validation_split=0.2, verbose=1,
                        callbacks=[early_stop],
                        shuffle=False)
```

```python
cwd = os.getcwd()
model_path = os.path.join(cwd,'LSTM_model')
lstm_model.save(os.path.join(model_path, 'LSTM_model.h5'))
```
